# Question 5
*************************************************
Like/Unlike posts
Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

steps)  
1)first went to the profile url of dilsefoodie    
2) clicked on the first post  
3)looped 30 times and done following:    
    i)if the ❤️ logo svg has aria lable==like then printed already liked     
    ii)else liked it   
    iii) clicked on the next icon if it was present or break the loop if it was not presnt  

*******************************************************

```python
account="dilsefoodie"
from bs4 import BeautifulSoup as bs
if driver.current_url !="https://www.instagram.com/{}/".format(account):
    driver.get("https://www.instagram.com/{}/".format(account))
    wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div._aagw")))
ele=driver.find_element(By.CSS_SELECTOR,"div._aagw") 
ele.click()
for i in range(30):
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,"section._aamu._ae3_._ae47._ae48")))
    like_button=driver.find_element(By.CSS_SELECTOR,"section._aamu._ae3_._ae47._ae48>span._aamw")
    like_unlike=bs(like_button.get_attribute("innerHTML"),"html.parser")
    if(like_unlike.find("svg").attrs.get("aria-label")=="Like"):
        like_button.click()
    else:
        print("already_liked")
    try:
        next=driver.find_element(By.CSS_SELECTOR,"div._aaqg._aaqh")
        next.click()
    except:
        break
        
```
